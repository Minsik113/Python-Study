```python
iris_dataset = load_iris()
iris_data = iris_dataset.data

# 붓꽃 데이터 세트를 자세히 보기 위해 DataFrame으로 변환하자
iris_df = pd.DataFrame(data=iris_data
                      ,columns=iris_dataset.feature_names)
iris_df.head(3)
```

```python
iris_df['label'] = iris_dataset.target
iris_df.head(3)
```

```python
# 유사 난수 생성기를 사용해 데이터셋을 무작위로 섞는다.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
                train_test_split(iris_dataset['data']
                                 ,iris_dataset['target']
                                 ,random_state=0)
# 75% 와 25%로 나뉘어졌다.
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(X_test.shape)
```

```python
import pandas as pd
iris_dataframe = pd.DataFrame(X_train
                              , columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe
                          ,c=y_train
                          ,figsize=(15,15)
                          ,marker='o'
                          ,hist_kwds={'bins':20}
                          ,s=60
                          ,alpha=.8
                          ,cmap=mglearn.cm3)
```

# 2. kNN최근접 이웃 알고리즘

- 유클리디안
- 이해하기 쉬운 모델이지만, 예측이 느리고 많은 특성을 처리하기 부족하다

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
# 훈련데이터셋으로부터 모델을 만들려면 knn객체의 fit메서드를 사용한다.
knn.fit(X_train, y_train)
```

### 2-1. 예측하기

```python
import numpy as np
X_new = np.array([[5, 2.9, 1, 0.2]])
print(X_new.shape)
```

```python
# 예측하기
prediction = knn.predict(X_new)
print(f'예측 {prediction}')
print(f'예측한 타깃의 이름 {iris_dataset["target_names"][prediction]}')
```

### 2-2.모델 평가하기

```python
# 모델 평가하기
y_pred = knn.predict(X_test)
print(f'테스트 세트에 대한 예측값  {y_pred}')
print(f'테스트 세트에 대한 실제값  {y_test}')
```

```python
print(f'테스트 세트의 정확도: {np.mean(y_pred==y_test):.2f}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
# 훈련데이터셋으로부터 모델을 만들려면 knn객체의 fit메서드를 사용한다.
knn.fit(X_train, y_train)
print(f"테스트 세트의 정확도 {knn.score(X_test,y_test)}")
```
