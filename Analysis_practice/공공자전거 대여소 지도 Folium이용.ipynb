```python
df_stop_total=df_stop.groupby('구분')['거치대수'].sum().to_frame()
df_stop_total.head()
```

```python
map=folium.Map(location=[37.5202, 126.975]
               ,zoom_start=10.5)
mc=fp.MarkerCluster(control=False)
map.add_child(mc)

geo_path='C:\\Users\\user\\Downloads\\DataScience-master\\DataScience-master\\data\\02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path,encoding='utf-8'))

map.choropleth(geo_data=geo_str
               ,data=df_stop_total #
               ,columns=[df_stop_total.index, '거치대수'] #
               ,fill_Colot='YlGnBu'
               ,key_on='feature.id')

for n in df_stop.index:
    if pd.notnull(df_stop['위도'][n]):
        folium.Marker([df_stop['위도'][n], df_stop['경도'][n]]
                       ,icon=folium.Icon(color='blue',icon='ok')
                       ,popup='%s<br>%d'%(df_stop['대여소명'][n],df_stop['거치대수'][n])
                       ).add_to(mc)
map                       
```
