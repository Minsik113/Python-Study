```python
# 3) 1년후, 2022.12.31 까지의 정보를 예측해보자
import pandas as pd

spy_trunc = spy[:'2022-01-01']
df = pd.DataFrame({
    'ds':spy_trunc.index
    ,'y':spy_trunc['Close']
})
df.reset_index(inplace=True)
del df['Date']
df
```

```python
# 주기성이 일단위로 있다고 알려준다
from fbprophet import Prophet
m = Prophet(daily_seasonality=True)
m.fit(df)

# 1년(365일) 후 주가를 예측하겠다
future = m.make_future_dataframe(periods=365)

# 예측 시각화
forecast = m.predict(future)
m.plot(forecast)
```

```python
# 4) 700일 간의 데이터를 예측
future = m.make_future_dataframe(periods=700)
# 예측 시각화
forecast = m.predict(future)
m.plot(forecast)
```
