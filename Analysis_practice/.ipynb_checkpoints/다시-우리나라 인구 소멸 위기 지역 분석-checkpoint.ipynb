## 3. 데이터 분석

### 3-0. fillna사용해서 nan처리
- df.fillna(0) # 결측값을 특정값으로 채우기
- df.fillna('missing)
- fillna(method='fill' or 'pad') 결측값을 앞 방향 혹은 뒷 방향으로 채우기
- fillna(method='bfill' or 'backfill')

```python
population.fillna(method='pad', inplace=True)
population.head(3)
```

```python
population.rename(columns = {
                    "행정구역(동읍면)별(1)" : '광역시도'
                    ,"행정구역(동읍면)별(2)": '시도'
                    ,'계' : '인구수'}
                 ,inplace=True)
population = population[(population['시도'] != '소계')]
population.head(3)
```

* **항목** 이라는 컬럼의 내용이 각 행정구역마다 총인구수, 남자인구수, 여자인구수 로 나눠있는 것을 알 수 있다.

```python
population.rename(columns = {"항목":'구분'}, inplace=True)

population.loc[population['구분'] == '총인구수 (명)', '구분'] = '합계' # 긴 이름을 짧게
population.loc[population['구분'] == '남자인구수 (명)', '구분'] = '남자'
population.loc[population['구분'] == '여자인구수 (명)', '구분'] = '여자'
population.head(3)
```

```python
population['20-39세'] = population['20 - 24세'] + population['25 - 29세'] + \
                        population['30 - 34세'] + population['35 - 39세']

population['65세이상'] = population['65 - 69세'] + population['70 - 74세'] + \
                        population['75 - 79세'] + population['80 - 84세'] + \
                        population['85 - 89세'] + population['90 - 94세'] + \
                        population['95 - 99세'] + population['100+'] 
population.head(3)
```

### 3-1. pivot_table생성
pivot_table을이용하여 광역시도, 시도를 index로 두고, 구분으로 세로를 첫번째 컬럼으로 잡고, value에 인구수, 20~39세, 65세이상으로 정리해둔다

```python
pop = pd.pivot_table(population
                    ,index=['광역시도','시도']
                    ,columns=['구분']
                    ,values=['인구수','20-39세','65세이상'])
pop
```

### 3-2. 인구소멸 비율 계산해서 추가
* 소멸비율이라는 컬럼에 인구소멸위기지역을 계산하기 위한 식을 적용한다.
* 비율이1보다 작으면 인구소멸 위기지역으로 볼 수 있따.

```python
pop['소멸비율'] = pop['20-39세','여자'] / (pop['65세이상','합계']/2) # 남자들어가있어서 2나눔
pop.head()
```

```python
pop['소멸위기지역'] = pop['소멸비율'] < 1.0
pop.head()
```

```python
pop[pop['소멸위기지역']==True].index.get_level_values('시도')
```

```python
pop[pop['소멸위기지역']==True].index.get_level_values(1)
```

```python
pop.reset_index(inplace=True)
pop.head()
```

### 3-3. 이중 column을 해제하기 위해 두 컬럼 제목을 합쳐 다시 지정한다

```python
print(pop.columns.get_level_values(0)) # 0 level
print(pop.columns.get_level_values(1)) # 1 level
```

```python
tmp_columns = [pop.columns.get_level_values(0)[n] + \
               pop.columns.get_level_values(1)[n]
               for n in range(len(pop.columns.get_level_values(0)))]
tmp_columns
```

```python
pop.columns = tmp_columns
pop.head()
```

```python
pop.info()
```

### 3-4. 지도 시각화를 위해 지역별 고유 ID 만들기

#### 인천, 부산 등 광역시 --> 자치구
#### 안양시, 수원시 --> 행정구

```python
si_name = [None] * len(pop)

tmp_gu_dict = {'수원' : ['장안구','권선구','팔달구','영통구']
               ,'성남':['수정구','중원구','분당구']
               ,'안양':['만안구','동안구']
               ,'안산':['상록구','단원구']
               ,'고양':['덕양구','일산동구','일산서구']
               ,'용인':['처인구','기흥구','수지구']
               ,'청주':['상당구','서원구','흥덕구','청원구']
               ,'천안':['동남구','서북구']
               ,'전주':['완산구','덕진구']
               ,'포항':['남구','북구']
               ,'창원':['의창구','성산구','진해구','마산합포구','마산회원구']
               ,'부천':['오정구','원미구','소사구']}
tmp_gu_dict
```

```python
for n in pop.index:
    if pop['광역시도'][n][-3:] not in ['광역시','특별시','자치시']:
        if pop['시도'][n][:-1]=='고성' and pop['광역시도'][n][:-1]=='강원도':
            si_name[n] = '고성(강원)'
        elif pop['시도'][n][:-1]=='고성' and pop['광역시도'][n][:-1]=='경상남도':
            si_name[n] = '고성(경남)'
        else:
            si_name[n] = pop['시도'][n][:-1]
        
        for keys, values in tmp_gu_dict.items():
            if pop['시도'][n] in values:
                if len(pop['시도'][n]) == 2: # '포항' : ['남구','북구']
                    si_name[n] = keys + ' ' + pop['시도'][n] #포항 남구
                elif pop['시도'][n] in ['마산함포구','마산회원구']: # '창원' : ['마산함포구','마산회원구']
                    si_name[n] = keys +' ' + pop['시도'][n][2:-1] # 창원 함포
                else: # '안양' : ['만안구','동안구']
                    si_name[n] = keys +' ' + pop['시도'][n][:-1] #안양 만안
                    
    elif pop['광역시도'][n] == '세종특별자치시':
        si_name[n] = '세종'
    else:
        if len(pop['시도'][n])==2: # 서울특별시 중구
            si_name[n]=pop['광역시도'][n][:2] + ' ' + pop['시도'][n] # 서울 중구
        else:
            si_name[n]=pop['광역시도'][n][:2] + ' ' + pop['시도'][n][:-1] # 서울 구로
```
