```python
from tqdm import tqdm_notebook
import time

text_all = []
for i in tqdm_notebook(range(1,1000,10)):
    response = urlopen(url.format(num=1, key_word=urllib.parse.quote('원소주')))
                                
    soup = BeautifulSoup(response, 'html.parser')
    tmp = soup.find_all(class_='answer_area')

    for tmp_text in tmp:
        text_all.append(tmp_text.text) # text부분만 뗴어온다
    time.sleep(0.5)
```
