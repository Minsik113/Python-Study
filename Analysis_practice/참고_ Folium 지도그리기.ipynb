```python
import pandas as pd
df = pd.read_csv('../data/03. best_sandwiches_chicago22.csv', index_col=0)
df.head(5)
```

```python
import googlemaps
gmaps_key = "~~" # 2장에서 획득한 자신의 key를 사용합니다.
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
import numpy as np
from tqdm import tqdm_notebook
lat = []
lng = []

for n in tqdm_notebook(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n]+', '+'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
df['lat'] = lat
df['lng'] = lng
df.head()
```

```python
import folium
mapping = folium.Map(location=[df['lat'].mean(),df['lng'].mean()]
                     ,zoom_start=11
                    )
folium.Marker([df['lat'].mean(), df['lng'].mean()]
             ,popup='center'
             ).add_to(mapping)
mapping
```

```python
map = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],zoom_start=11)

for i in df.index:
    if df['Address'][i] != 'Multiple':
        folium.Marker([df['lat'][i], df['lng'][i]]
                     ,popup=df['Cafe'][i]
                     ).add_to(map)
map
```

```python
from folium.plugins import MarkerCluster
mc = MarkerCluster()

map = folium.Map(location=[df['lat'].mean(), df['lng'].mean()]
                ,zoom_start=11)
for i in df.index:
    if df['Address'][i] != 'Multiple':
        mc.add_child(folium.Marker(location=[df['lat'][i], df['lng'][i]]
                                  , popup=df['Cafe'][i]))
        
map.add_child(mc)
map
```
