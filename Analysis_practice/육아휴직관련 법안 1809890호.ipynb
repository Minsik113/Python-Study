# 3. Naive bayes Classifier의 이해 - 한글

```python
from konlpy.tag import Okt

pos_tagger = Okt()
train = [('메리가 좋아', 'pos'), 
         ('고양이도 좋아', 'pos'),
         ('난 수업이 지루해', 'neg'),
         ('메리는 이쁜 고양이야', 'pos'),
         ('난 마치고 메리랑 놀거야', 'pos')]
```

```python
# 말뭉치
all_words = set(word.lower() for sentence in train
                               for word in word_tokenize(sentence[0]))
all_words
```

```python
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1])
                                                for x in train]
t
```

```python
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```python
test_sentence = '난 수업이 마치면 메리랑 놀거야'
test_sent_features = {word.lower():
                         (word in word_tokenize(test_sentence.lower()))
                                 for word in all_words}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```

```python
from konlpy.tag import Okt
twitter = Okt()

print(twitter.pos('이것도 재미있습니당 ㅋㅋ'))
print('-------')
print(twitter.pos('이것도 재미있습니당 ㅋㅋ', norm=True)) # norm = 정규화
print('-------')
print(twitter.pos('이것도 재미있습니당 ㅋㅋ', norm=True, stem=True)) # stem = 원형
```

```python
def tokenize(doc):
    return ['/'.join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)]

train_docs = [(tokenize(row[0]), row[1]) for row in train]
train_docs
```

```python
tokens = [t for d in train_docs for t in d[0]]
tokens
```

```python
def term_exists(doc):
    return {word: (word in set(doc)) for word in tokens}

# '메리가 좋아', '고양이도좋아'
# 토큰라이징한 후 문장이 말뭉치에 있는지 확인
train_xy = [(term_exists(d), c) for d,c in train_docs]
train_xy
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_xy)
```

```python
test_sentence = [("난 수업이 마치면 메리랑 놀거야")]
test_docs = pos_tagger.pos(test_sentence[0])
test_docs
```

```python
classifier.show_most_informative_features()
```

```python
test_sent_features = {word: (word in tokens) for word in test_docs}
test_sent_features
```

```python
classifier.classify(test_sent_features)
```

# 4. 문장의 유사도 측정


```python
from sklearn.feature_extraction.text import CountVectorizer

# 단어장에 포함되기 위한 최소 빈도. min_df=1 -> default = 1이라 생랼가능
vectorizer = CountVectorizer(min_df=1) 

contents = ['메리랑 놀러가고 싶지만 바쁜데 어떻하죠?',
           '메리는 공원에서 산책하고 노는 것을 싫어해요',
           '메리는 공원에서 노는 것도 싫어해요. 이상해요.',
           '먼 곳으로 여행을 떠나고 싶은데 너무 바빠서 그러질 못하고 있어요']

x = vectorizer.fit_transform(contents)
vectorizer.get_feature_names()
```

```python
'''
1열보면 1로 되어있는 단어들이 '놀러가고' '메리랑' '바쁜데' '싶지만' '어떻하죠'이다.

선언한 문장이 4개니까 열이4개 나오는거고, vector화시킨 vectorizer.get_feature_names()에서
속하는걸 1로 표현
'''
x.toarray().transpose()
```

```python
x = vectorizer.fit_transform(contents)
num_samples, numfeatures = x.shape
num_samples, numfeatures
```

```python
new_post = ['메리랑 공원에서 산책하고 놀고 싶어요']
new_post_vec = vectorizer.transform(new_post)
new_post_vec.toarray() # 4번쩨-'공원에서' 11번째-'메리랑' 15번째-'산책하고'
```

```python
# linalg.norm() 명령으로 벡터의 길이를 계산
# new_post = ['메리랑 공원에서 산책하고 놀고 싶어요']
import scipy as sp
def dist_raw(v1, v2):
    delta = v1 - v2
    return sp.linalg.norm(delta.toarray())


best_doc = None
best_dist = 65535
best_i = None
 # new_post_vec => '공원에서' 메리랑' '산책하고'
for i in range(0, num_samples): # num_samples = 4
    post_vec = x.getrow(i)
    
    # new_post_vec와 post_vec 문장사이의 거리 측정
    d = dist_raw(post_vec, new_post_vec) 
    print("== Post %i with dist=%.2f : %s" %(i,d,contents[i]))
    if d < best_dist:
        best_dist = d
        best_i = i
```
